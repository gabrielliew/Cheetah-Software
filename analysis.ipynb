```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy import signal
from scipy.signal import find_peaks
from scipy import signal
from scipy.interpolate import CubicSpline
import mpl_axes_aligner 
import seaborn as sns
sns.set_theme(style="darkgrid")

%matplotlib widget
```

```python
def draw_one_cycle_graph_for_analysis(graph_name,data_input,joint):
    sns.set(rc = {'figure.figsize':(15, 7)})
    peak, _ = find_peaks(data_input.qd01, height=2)
    spacing_list=[]
    for i in range(len(peak)-1):
        spacing_list.append(peak[i+1]-peak[i])
    spacing=int(np.mean(spacing_list))
    dataframe_gait = pd.DataFrame({"gait_progression":[],
                                   "q":[],
                                   "qd":[],
                                   "torque":[]})

    for i in range(len(peak)-1):
        # if(peak[i+1]-peak[i])!=spacing:
            # f = CubicSpline(list(range(0,peak[i+1]-peak[i])),data_input.qd01[peak[i]:peak[i+1]].tolist(), bc_type='natural')
            # x_new = np.linspace(0, 1, spacing)
            # y_new = f(x_new) #not doing shit
            # holder.append(y_new.tolist())
        # else:
         if(peak[i+1]-peak[i])==spacing:
            if joint=="ankle":
                temp = pd.DataFrame({"gait_progression":np.linspace(0, 1,spacing).tolist(),
                                     "q":data_input.q02[peak[i]:peak[i+1]].tolist(),
                                     "qd":data_input.qd02[peak[i]:peak[i+1]].tolist(),
                                     "torque":data_input.t02[peak[i]:peak[i+1]].tolist()})
                dataframe_gait=pd.concat([dataframe_gait, temp])
            elif joint=="knee":
                temp = pd.DataFrame({"gait_progression":np.linspace(0, 1,spacing).tolist(),
                                     "q":data_input.q01[peak[i]:peak[i+1]].tolist(),
                                     "qd":data_input.qd01[peak[i]:peak[i+1]].tolist(),
                                     "torque":data_input.t01[peak[i]:peak[i+1]].tolist()})
                dataframe_gait=pd.concat([dataframe_gait, temp])

    # plt.close()
    # plt.figure(figsize=(20,10))
    fig, ax = plt.subplots()
    # dataframe_gait
    sns.lineplot(
        data=dataframe_gait,
        x="gait_progression", y="torque", errorbar="sd",ax=ax,color='red'
    )
    ax2 = ax.twinx()
    sns.lineplot(
        data=dataframe_gait, 
        x="gait_progression", y="qd", errorbar="sd",ax=ax2,color='green'
    )
    mpl_axes_aligner.align.yaxes(ax, 0, ax2, 0, 0.5)
    plt.axhline(y=0.0, color='black', linestyle='-')
    plt.title(graph_name) 
    plt.show()
    
def draw_one_cycle_graph_for_analysis_per_leg(prefix,data):
    draw_one_cycle_graph_for_analysis(prefix+"_knee",data,"knee")
    draw_one_cycle_graph_for_analysis(prefix+"_ankle",data,"ankle")

# sns.relplot(
#     data=dataframe_gait, kind="line",
#     x="gait_progression", y="q", errorbar="sd",
# )
```
